```python
import numpy as np

# Original permeability values (mD)
values = np.array([50, 80, 100, 90, 140, 30])

# Step 1: Calculate original mean and std dev
mean_original = np.mean(values)
std_original = np.std(values)

# Target normal distribution parameters
mean_target = 1.0
std_target = 4.0

# Step 2: Normalize and transform
z_scores = (values - mean_original) / std_original
transformed_values = z_scores * std_target + mean_target

# Output results
print("Original values:", values)
print("Original mean:", mean_original)
print("Original std dev:", std_original)
print("Z-scores:", z_scores)
print("Transformed values:", transformed_values)
```

```python
import numpy as np

# Parameters
a = 1
m = 2
n = 2
Rw_Rt = 0.01
porosity_mean = 0.2
porosity_std = 0.02
num_samples = 100000  # Large number for accurate estimation

# Generate random porosity values
porosity_samples = np.random.normal(porosity_mean, porosity_std, num_samples)

# Ensure porosity stays within physical bounds (0 < φ < 1)
porosity_samples = np.clip(porosity_samples, 0.01, 0.99)

# Calculate Sw for each sample
Sw_samples = np.sqrt(Rw_Rt * a / (porosity_samples**m))

# Calculate statistics
Sw_mean = np.mean(Sw_samples)
Sw_variance = np.var(Sw_samples)

print(f"Mean of Sw: {Sw_mean:.4f}")
print(f"Variance of Sw: {Sw_variance:.6f}")
```
